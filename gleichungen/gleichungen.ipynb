#### A2020

```python
var('x', domain='real')
f(x) = (x+4)^2
h(x) = 1/5*(x^2-4)
solve(f,x), solve(h,x)
```

```python
solve(f(x)==h(x),x)
```

```python
h(-3),h(-7)
```

```python
plot([f(x),h(x)], (x,-8,3),ymin=-1,ymax=20,)
```

```python
solve(h(x)<=f(x),x)
```

```python
solve([h(x)>=0, f(x)>=0, x+4>=0, h(x)<=f(x)],x)
```
